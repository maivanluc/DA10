```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd

def f_read(name,comp_name):
    stock = pd.read_csv(name)
    stock['<DTYYYYMMDD>'] = pd.to_datetime(stock['<DTYYYYMMDD>'], format = '%Y%m%d')
    stock.set_index(['<DTYYYYMMDD>'], inplace=True)
    stock['<Open>'].plot(color="green")
    plt.title('Biểu đồ '+ comp_name)
    plt.xlabel('Year')
    plt.ylabel('Open')
    plt.show()


f_read("excel_acb.csv","ACB")     
f_read("excel_dxg.csv","DXG") 
f_read("excel_flc.csv","FLC") 
f_read("excel_fpt.csv","FPT") 
f_read("excel_vjc.csv","VJC") 
f_read("excel_vnm.csv","VNM") 
```

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd

def f_run():
    fig, ax = plt.subplots(figsize=(5, 5), layout='constrained')
    ACB = pd.read_csv("excel_acb.csv")
    ACB['<DTYYYYMMDD>'] = pd.to_datetime(ACB['<DTYYYYMMDD>'], format = '%Y%m%d')
    ACB.set_index(['<DTYYYYMMDD>'], inplace=True)
    
    DXG = pd.read_csv("excel_dxg.csv")
    DXG['<DTYYYYMMDD>'] = pd.to_datetime(DXG['<DTYYYYMMDD>'], format = '%Y%m%d')
    DXG.set_index(['<DTYYYYMMDD>'], inplace=True)
    
    FLC = pd.read_csv("excel_flc.csv")
    FLC['<DTYYYYMMDD>'] = pd.to_datetime(FLC['<DTYYYYMMDD>'], format = '%Y%m%d')
    FLC.set_index(['<DTYYYYMMDD>'], inplace=True)
    
    FPT = pd.read_csv("excel_fpt.csv")
    FPT['<DTYYYYMMDD>'] = pd.to_datetime(FPT['<DTYYYYMMDD>'], format = '%Y%m%d')
    FPT.set_index(['<DTYYYYMMDD>'], inplace=True)
    
    VJC = pd.read_csv("excel_vjc.csv")
    VJC['<DTYYYYMMDD>'] = pd.to_datetime(VJC['<DTYYYYMMDD>'], format = '%Y%m%d')
    VJC.set_index(['<DTYYYYMMDD>'], inplace=True)
    
    VNM = pd.read_csv("excel_vnm.csv")
    VNM['<DTYYYYMMDD>'] = pd.to_datetime(VNM['<DTYYYYMMDD>'], format = '%Y%m%d')
    VNM.set_index(['<DTYYYYMMDD>'], inplace=True)
       
  
    ax.plot(ACB['<Open>'], label='ACB') 
    ax.plot(DXG['<Open>'], label='DXG') 
    ax.plot(FLC['<Open>'], label='FLC') 
    ax.plot(FPT['<Open>'], label='FPT') 
    ax.plot(VJC['<Open>'], label='VJC') 
    ax.plot(VNM['<Open>'], label='VNM') 
    ax.legend() 
    plt.title('Biểu đồ các mã chứng khoán')
    plt.xlabel('Year')
    plt.ylabel('Open Value')
    plt.show()

f_run()
```
