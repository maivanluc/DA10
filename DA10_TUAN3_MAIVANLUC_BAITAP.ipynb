```python
def kiem_tra( x ):
    if x < 2 :
        return 0
    for i in range(2, x//2 + 1):
        if x%i==0:
            return 0
    return 1

so_luong = int(input("Nhập số lượng số trong list: "))

list_so = []
x = False

for i in range(0,so_luong):
    
    while not x :
        so_tn = (input("Nhập doanh số thứ {}: ".format(i+1)))
        if not so_tn.isnumeric() :
            print("{} Phải là có giá trị số ".format(so_tn))     
        elif int(so_tn) < 0:
            print(" {} phải lớn hơn = 0 ".format(so_tn))
        else:
            x =True
            
    list_so.append(int(so_tn))
    x = False
 
print(f"List dãy số : {list_so}")
list_kq =[]
for i in range(0,len(list_so)):
    check = kiem_tra(list_so[i])
    if check == 0:
        s=(" {} không là số nguyên tố ".format(list_so[i]))
    else:
        s=(" {}  là số nguyên tố ".format(list_so[i]))
    list_kq.append(s)
    print(s)
    
with open("so_nguyen_to.txt", mode='w', encoding="utf-8") as f:
    for kq in list_kq:
        f.write(kq)   
        f.write('\n')
        
```

```python
def tinh_thue(luong):
    if luong <= 5000000 :
        return 0.05*luong
    elif luong > 5000000 and luong <= 100000000 :
        return 0.1 *luong-250000
    elif luong > 5000000 and luong <= 100000000 :
        return 0.1 *luong-250000
    elif luong > 10000000 and luong <= 180000000 :
        return 0.15 *luong-750000
    elif luong > 18000000 and luong <= 320000000 :
        return 0.2 *luong-1650000
    elif luong > 32000000 and luong <= 520000000 :
        return 0.25 *luong-3250000
    elif luong > 52000000 and luong <= 800000000 :
        return 0.3 *luong-5850000
    else:
        return 0.35 *luong -9850000

def bao_hiem(luong):
    return 0.105*luong


def nhap_danh_sach_nhan_vien():
    danh_sach_nhan_vien = []
    so_luong_nhan_vien = int(input("Nhập số lượng nhân viên: "))

    for i in range(so_luong_nhan_vien):
        print(f"\nNhập thông tin cho nhân viên thứ {i+1}:")
        
        luong = int(input("Lương: "))
        phu_cap = int(input("Phụ cấp: "))

        nhan_vien = {"luong": luong, "phucap": phu_cap, "tong": luong + phu_cap}
        danh_sach_nhan_vien.append(nhan_vien)

    return danh_sach_nhan_vien

def tinh_luong_bao_hiem(danh_sach_nhan_vien):
    nhan_vien_bh_phucap = []

    for nhan_vien in danh_sach_nhan_vien:
        thue =tinh_thue(nhan_vien["tong"])
        bh =bao_hiem(nhan_vien["tong"])
        nhan_vien_total={"luong": nhan_vien["luong"], "phu_cap": nhan_vien["phucap"], "tong_tien":nhan_vien["tong"],"thue":thue,"bh":bh}
        nhan_vien_bh_phucap.append(nhan_vien_total)

    return nhan_vien_bh_phucap

danh_sach_nhan_vien = nhap_danh_sach_nhan_vien()
thue_bh = tinh_luong_bao_hiem(danh_sach_nhan_vien)
print("\n==================================")
print("\nDanh sách Lương Phụ cấp từng nhân viên")
with open("danh_sach_luong_phu_cap.txt", mode='w', encoding="utf-8") as f:
    for nhan_vien in thue_bh:
        i = 1
        s=(f"Nhân viên số:{i}")
        print(s)
        f.write(s)
        f.write('\n')
        t = nhan_vien["luong"]
        s=(f"Lương: {t}")
        print(s)
        f.write(s)
        f.write('\n')
        t =nhan_vien["phu_cap"]
        s=(f"Phụ cấp:  {t}")
        print(s)
        f.write(s)
        f.write('\n')
        t=nhan_vien["tong_tien"]
        s=(f"Tổng thu nhập:  {t}")
        print(s)
        f.write(s)
        f.write('\n')
        t=nhan_vien["thue"]
        s=(f"Tiền chịu thuế: {t}")
        print(s)
        f.write(s)
        f.write('\n')
        t=nhan_vien["bh"]
        s=(f"Tiền chịu bảo hiểm: {t}")
        print(s)
        f.write(s)
        f.write('\n')
        print("---")
        i = i + 1
```

```python
import csv
import statistics

ds_tp = []
ds_gioitinh =[]
ds_tongdon =[]
ds_rating =[]
def tso_tsuat_khach_hang_tp(ds_tpp):
    #ds_ten_tp = list(set(ds_tp))
    ds_tp1 = []
    for tp in ds_tpp:
        if tp.upper() =="TP HCM" or tp.upper() =="HÀ NỘI" or tp.upper() =="ĐÀ NẴNG"  or tp.upper() =="NAM"  or tp.upper() =="NỮ":
            ds_tp1.append(tp)
    

    unique_tp = list(set(ds_tp1))
    prequency=[]
    for t in unique_tp:
        freq = ds_tpp.count(t)
        prequency.append((t,freq))
    
    total_count = len(ds_tpp)
    relative_prequency=[]
    for t,freq in prequency:
        req_fred = freq/total_count
        relative_prequency.append((t,req_fred))
    
    for t,freq in prequency:
        print(f"Tần số khách hàng  {t} là {freq}")
    for t,refreq in relative_prequency:
        print(f"Tần suất khách hàng {t} : {refreq:.2%}")  
        

with open("sale_data_vn.csv", mode='r', encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)
    print(header)
    for row in reader:
        ds_tp.append(row[2])
        ds_gioitinh.append(row[4])
        ds_tongdon.append(row[10])
        ds_rating.append(row[14])
   # print(ds_gioitinh)    
    

print("-------Thành phố-------")
tso_tsuat_khach_hang_tp(ds_tp)
print("-------Giới tính-------")
tso_tsuat_khach_hang_tp(ds_gioitinh)

ds_rating_float = [float(x) for x in ds_rating]
ds_tongdon_float = [float(x) for x in ds_tongdon]

x = statistics.mean(ds_tongdon_float)
y =  statistics.median(ds_tongdon_float)

print("Tổng đơn - Mean :", x)
print("Tổng đơn - Median :", y)
x = statistics.mean(ds_rating_float)
y =  statistics.median(ds_rating_float)

print("Rating - Mean :", x)
print("Rating - Median :", y)
```
